### Example Task 4 - Multiple pages and elements

http://books.toscrape.com/catalogue/category/books_1/index.html
    
get a title of every book that has a two star rating from this link

```python
# understnd the procidure from swiping pages

# p1 - http://books.toscrape.com/catalogue/category/books_1/page-1.html
# p2 - http://books.toscrape.com/catalogue/category/books_1/page-2.html
# p3 - http://books.toscrape.com/catalogue/category/books_1/page-3.html
```

```python
# copy and format the link the {} instead of page number

base_url = "http://books.toscrape.com/catalogue/category/books_1/page-{}.html"
```

```python
# create a place holder for numbers for looping

page_num = 12
base_url.format(page_num)
```

```python
# we need to understand what tag and class represent that rating
# inspect the picture to get an idea
```

```python
# create a request

res_link = requests.get(base_url.format(1))
```

```python
soup = bs4.BeautifulSoup(res_link.text, "lxml")
```

```python
# the soup of code of page 1

soup
```

```python
# now lets select the class call (use .classname on classes html)
# "." is equal to "class =" name_of_class

products = soup.select(".product_pod")
products
```

```python
# is showing 20 items is a list
# it makes sence since it shows the number of books a page represents

len(products)
```

```python
# the 1st class call on the page
# basically the first book on page 1

example = products[0]
```

```python
# html code for book 1 repsentation

example
```

```python
# 1 way to search is risky (may not show always)

# 1-set this to a string "example sup to a str"
# 2-check if the "star-rating Two" is in the str
# 3-get boolean value to work with
```

```python
"start-rating Two" in str(example)
```

```python
# 2 way is more preceise: will be checking the classes presence
```

```python
# since it is a special obejct, try to find the necessary class in it
# use . instead of whitespaces and . to indicate a class at first

example.select(".star-rating.Three")
```

```python
# check if the returned list is empty
# like in example below
# remepber example is the 1st book on p1

example.select(".star-rating.Two")==[]
```

```python
# how to return a title
# check the code for a book title class or call

# kept in between <a> </a>
# we need the second link

example.select("a")[1]["title"]
```

```python
# Check if something is 2-stars (string call in, example.select(rating))
# example.select("a")[1]["title"] - book title
# combine with for loop
```

```python
two_star_title = []

for n in range(1,51):
    
    scrape_url = base_url.format(n)
    res = requests.get(scrape_url)
    
    soup = bs4.BeautifulSoup(res.text, "lxml")
    books = soup.select(".product_pod")
    
    for book in books:
        if len(book.select(".star-rating.Two")) != 0:
            book_title = book.select("a")[1]["title"]
            two_star_title.append(book_title)
```
